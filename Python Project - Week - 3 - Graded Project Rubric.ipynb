# 7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on..... (5 marks)

# 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot



```python
correlation = df['view_count'].corr(df['dislikes'])
correlation
```

```python
print("Correlation between views and dislikes:", correlation)
```

```python
# Analyze the significance of the correlation

if correlation > 0:
    print("There is a positive correlation between views and dislikes.")
elif correlation < 0:
    print("There is a negative correlation between views and dislikes.")
else:
    print("There is no linear correlation between views and dislikes.")
```
